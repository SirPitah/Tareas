```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#crear array angulos y sin(x)
x=np.linspace(-4*np.pi,4*np.pi,500)
fx=np.sin(x)

#crear array con aproximacion
tx=[]
N = 10  #hasta donde truncar
for aux_x in x: #por cada valor de x se hace la suma de taylor hasta N
	aux_y=0
	for n in range(0,N+1):
		aux_y += (-1)**(n)*aux_x**(2*n+1)/np.math.factorial(2*n+1)
	tx.append(aux_y)

tx_array=np.array(tx)
dif=np.abs(fx-np.array(tx))  #valor absoluto de la diferencia al aproximar

#encontrar rango de tolerancia
eps=0.10  #10% de la amplitud maxima sen(x)
xmin,xmax =0,0
#loop para encontrar los valores de x donde la diferencia pasa por el umbral de la tolerancia
for n in range(len(dif)-1):
	if (dif[n]>=eps) & (dif[n+1]<eps):
		xmin=x[n]
	elif (dif[n]<=eps) & (dif[n+1]>eps):
		xmax=x[n]

#graficar la funcion, aproximacion y el avalor absoluto de la diferencia
fig=plt.figure()
ax=fig.add_subplot()

#graficar lo importante
ax.plot(x,fx,label='sin(x)')
ax.plot(x,tx_array,color='cyan',linestyle='--',label='Taylor truncado a 10')

#graficar lo bonito
#ax.plot(x,dif,color='green',linestyle=':',label='Resto')
ax.vlines([xmin,xmax],-1.5,1.5,linestyle='-',color='red',label='Rango aceptable')

ax.set_ylim(-1.5,1.5)
ax.set_xlim(-10,10)
plt.legend()
plt.title('Expansión de Taylor hasta {}-ésimo termino'.format(N))
plt.show()
```
